```python
params = {'limit': '24', 
          'offset': '0',
          'latitude': '53.348779',
          'longitude': '83.776282',
          'fastFilters': 'group',
          'sort': 'by_popularity',
          'cityId': '32',
          'cacheBreaker': '1663833370'}

url = 'https://api.delivery-club.ru/api1.3/screen/web/components'
#?limit=24&offset=0&latitude=53.348779&longitude=83.776282&fastFilters=group&sort=by_popularity&cityId=32&cacheBreaker=1663833370'
```

```python
response = requests.get(url=url, params=params )
```

```python
k = 0
for vendor in vendors:
  for vendor1 in vendor["vendor"]['delivery']['priceIntervals']:
    if (vendor1['deliveryPrice'] == 0) & (vendor1['minOrderTotal'] == 0):
      print(vendor["vendor"]['name'], vendor1['deliveryPrice'])
      k = k + 1
        
```

```python
print(f'Доля (в %) ресторанов с бесплатной доставкой  = {k / len(vendors) * 100:.2f}%')
```
